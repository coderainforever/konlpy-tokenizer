```python
from konlpy.tag import Okt
from konlpy.utils import pprint
```

```python
text = input("enter text to analyze: ")
okt = Okt()
pprint(okt.pos(text, norm=True, stem=True, join=True))
```

```python
import pandas
from pandas import DataFrame
```

```python
# Use the okt.pos function and make a dataframe
ttrans = (okt.pos(text, norm=True, stem=True, join=True))
koreanlist = DataFrame(ttrans, columns=['Korean'])
```

```python
koreanlist
```

```python
koreanlist.info()
```

```python
# remove punctuation and sort based on word type
koreanlist['Word'] = koreanlist['Korean'].apply(lambda x: x.split('/')[0])
koreanlist['Type'] = koreanlist['Korean'].apply(lambda x: x.split('/')[1])
koreanlist.drop('Korean', axis=1, inplace=True)
finallist = koreanlist.drop_duplicates()
finallist.reset_index(drop=True, inplace=True)

finallist
```

```python
finallist.info()
```

```python
#counting number of words in each pos
count_pos = finallist.groupby('Type').count()
count_pos
```

```python
# optional save as a text file
finallist.to_csv('korean-tokenizer.txt', sep='\t', index=False)

# optional save as html
finallist.to_html('korean-NLP-tokenized.html')
```
